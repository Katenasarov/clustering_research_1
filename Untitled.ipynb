```python
# напишем функцию, для изменения формата данных
def modifications(df):
    '''
    Функция получает на вход датафрем,
    меняет формат данных,
    возвращает измененный датафрейм.
    '''
    for column in df.columns:
        column_type = df[column].dtypes
        if column_type == np.float64:
            df[column] = df[column].astype(np.float32)
    return df
```

## 6. FAISS

Создадим индекс для датасета base. Используем библиотеку FAISS с индексом IndexIVFFlat, с помощью которого векторы из base будут разбиты на кластеры с центроидой. Используется IndexFlatL2 - это простейший точный поиск ближайшего соседа с евклидовым расстоянием (норма L2). Метод .search() возвращает расстояния до k ближайших соседей и их индексы.

```python
# задаем размер индексных векторов
dims = base.shape[1]
print('Кол-во признаков в base', dims)
# задаем количество кластеров/центроид которое должно быть сформироано
n_cells = 30
# иницилиазируем квантователь - это индекс для присвоения векторов определенному кластеру
quantizer = faiss.IndexFlatL2(dims)
# формируем индекс
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```

```python
%%time
#обучаем индекс для создания кластеров
idx_l2.train(np.ascontiguousarray(base.values).astype('float32'))
print('Обучен индекс или не обучен', idx_l2.is_trained)
print('Кол-во векторов добавленных к индексу до момента добавления', idx_l2.ntotal) 
# добавляем в кластеры соответствующие векторы
idx_l2.add(np.ascontiguousarray(base.values).astype('float32'))
print('Кол-во векторов добавленных к индексу после добавления', idx_l2.ntotal)
```

```python
# индексы base переводим в счетчик, начинающийся с нуля
base_index = {k: v for k, v in enumerate(base.index.to_list())}
# выведем показатели индексов 10-ть штук
print('Индексы base\n', list(base_index.items())[0:10])
```

Выполним поиск пяти ближайщих соседей для тренировочных данных.

```python
# задаем кол-во индексов 5
candidate_number = 5
# задаем кол-во кластеров по которым необходимо пройтись
C_ = [5,10,20,30]
for c in C_:
    # запускаем таймер
    start = time.time()
    # задаем кол-во кластеров по которым необходимо пройтись
    idx_l2.nprobe = c
    # проводим поиск 5-ти индексов для тренировочной выборки, сделаем ограничение данные выберем первые 1000 значений, иначе ноут умирает
    r, idx = idx_l2.search(np.ascontiguousarray(train.values[:1000, :]).astype('float32'), candidate_number)
    # проверяем качество алгоритма accuracy@5 на тренирововной выборке
    acc = 0
    for target, el in zip(train_targets.values.tolist(), idx.tolist()):
        acc += int(target in [base_index[r] for r in el])
    # считаем сколько времени прошло
    loop_search_time = time.time() - start
    print('Кол-во кластеров =', c, f'Accuracy@{5} = {acc / len(idx):.1%}', f'Время:{loop_search_time}')
```

## 7. ML-модель

С целью улучшения показателей Faiss добавим ML-модель. Для этого из faiss заберем ее предсказания, и найдем 10 ближайших соседей, и с помощью модели классификации подберем ближайшего соседа.

```python
%%time
# задаем кол-во индексов 10
candidate_number = 10
# проводим поиск 10-ти индексов для тренировочной выборки, сделаем ограничение данные выберем первые 1000 значений, иначе ноут умирает
r, idx_10 = idx_l2.search(np.ascontiguousarray(train.values[:1000, :]).astype('float32'), candidate_number)
```

```python
# создадим свою функцию, которая обьеденит данные для того чтобы можно было обучить модель машинного обучения
def consolidation(index_faiss, target, base_index, data):
    '''
    Функция на вход принимает id товара,
    кандидаты в id товара,
    правильный ответ id товара из base, который максимально похож на него, 
    и признаки по которым можно обучать модель.
    Функция формирует датафрем из признаков, кандидата, верного ID и ID и base,
    Функция выделяет признак matching в случае если кандидат на верный id совпал с правильным id,
    удаляем ненужные признаки.
    '''
    df = pd.DataFrame(data=index_faiss, index=target.index)
    df = pd.melt(df.T)
    df.columns = ['id_query', 'idx_10']
    df['candidate'] = [base_index[number] for number in df['idx_10'].values]
    df = df.merge(target, left_on='id_query', how = 'left', right_index=True)
    df = df.merge(data, left_on='id_query', how = 'left', right_index=True)
    df = df.drop('idx_10', axis=1)
    df['matching'] = (df['candidate'] == df['target']).astype('int32')
    df.drop('target', axis=1, inplace=True)
    return df
```

```python
# сформируем тренирововчную выборку
Х_y_train = consolidation(index_faiss=idx_10, target=train_targets[:1000], base_index=base_index, data=train[:1000])
```

```python
# выведем первые пять строк, проверим как сработал код
Х_y_train.head(5)
```

```python
# удалим индексы, чтобы не возникало FutureWarning
Х_y_train = Х_y_train.reset_index(drop=True)
# разделяем данные на целевой признак - y и на входной признак X  в train (тренировочной выборке)
X_train = Х_y_train.drop(['matching', 'id_query'], axis=1)
y_train = Х_y_train['matching']
```

```python
# создадим списки с названиями признаков
# список с количественными признаками
num_columns = X_train.select_dtypes(include='number').columns.tolist()
# cписок с категориальными признаками
ohe_columns = X_train.select_dtypes(exclude='number').columns.tolist()
```

```python
%%time
# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline([('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', sparse=False, handle_unknown='ignore'))])

# создаём пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', 'passthrough', num_columns)]
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))])

param_grid = [
    # словарь для модели CatBoostClassifier
    {'models': [CatBoostClassifier(random_state=RANDOM_STATE, verbose=0)],
     'models__learning_rate': [0.01, 0.15],
     'models__iterations': [500, 1000],
     'models__max_depth': range(3, 6),
     'preprocessor__num': ['passthrough'],
     'preprocessor__ohe': [OrdinalEncoder(handle_unknown='ignore'), OneHotEncoder(handle_unknown='ignore')]
    },
    
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 8),
        'models__max_features': range(2, 8), 
        'preprocessor__num': ['passthrough'],
        'preprocessor__ohe': [OrdinalEncoder(handle_unknown='ignore'), OneHotEncoder(handle_unknown='ignore')]
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(1, 6),
        'preprocessor__num': ['passthrough'],
        'preprocessor__ohe': [OrdinalEncoder(handle_unknown='use_encoded_value'), OneHotEncoder(handle_unknown='use_encoded_value')]
    }
]

randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)

# проверяем работу модели
print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)
print('Лучшие параметры модели:', randomized_search.best_params_)
```

ROC-AUC (Area Under the Receiver Operating Characteristic Curve) — это метрика, которая измеряет качество бинарной классификации.

Значение AUC ближе к 1 указывает на хорошую модель, но наша метрика получилась очент и очень плохая, поэтому не будем ее тестировать на валидационной выборке. Остановимся на FAISS т.к. именно она показала вполне хороший результат.
